```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rlc_transfer_function_omega0_Q(omega, omega0, Q):
    """
    Calculates the complex transfer function H(ω) for a series RLC circuit
    in terms of resonant frequency ω₀ and quality factor Q.
    
    Parameters:
    omega0 (float): Resonant angular frequency ω₀ in rad/s
    Q (float): Quality factor
    omega (numpy array): Angular frequency ω in rad/s

    Returns:
    numpy array: Complex transfer function H(ω) values for each frequency in omega
    """
    # Calculate the complex transfer function
    H_omega = 1 / (1 - (omega / omega0)**2 + 1j * (omega / (omega0 * Q)))
    
    return H_omega


freqs = np.logspace(0, 6, 1000)
omega0 = 1e3
Q = 100
h_values = rlc_transfer_function_omega0_Q(freqs, omega0, Q)

fig, ax = plt.subplots()

ax.plot(freqs, 20*np.log10(np.abs(h_values)), label='Magnitude')
ax.set_xscale('log')

```
